```python
from collections import Counter
from datasets import load_dataset, DatasetDict
import random
import yaml
import os
```

```python
dataset = load_dataset("ds4sd/DocLayNet-v1.1")
```

```python

from datasets import load_dataset, DatasetDict
import random
def get_subset(split, size):
    if len(split) <= size:
        return split
    indices = random.sample(range(len(split)), size)
    return split.select(indices)

subset_dataset = DatasetDict({
    'train': get_subset(dataset['train'], 12000),
    'val': get_subset(dataset['val'], 4000),
    'test': get_subset(dataset['test'], 300),
})

subset_sizes = {split: len(subset_dataset[split]) for split in subset_dataset}
total_subset_size = sum(subset_sizes.values())

subset_percentages = {
    split: round(100 * count / total_subset_size, 2)
    for split, count in subset_sizes.items()
}

print("Subset sizes:", subset_sizes)
print("Subset percentages of total subset:", subset_percentages)
```

```python

from collections import Counter
from datasets import load_dataset


label_counter = Counter()

for split in subset_dataset:
    for sample in subset_dataset[split]:
        labels = sample["category_id"]
        # If multi-label (a list), extend the counter
        if isinstance(labels, list):
            label_counter.update(labels)
        else:
            label_counter[labels] += 1

# Print the results
for label_id, count in sorted(label_counter.items()):
    print(f"Class {label_id}: {count} instances")
```

```python

import torch
import torch.nn as nn
```

```python
num_classes = max(label_counter.keys()) + 1
subclass_counter = [label_counter[i] if i in label_counter else 0 for i in range(1,num_classes)]

print(subclass_counter)
```

```python

total = sum(subclass_counter)

subweights = [total / c for c in subclass_counter]

subweights = torch.tensor(subweights, dtype=torch.float)
subweights = subweights / subweights.sum()

criterion = nn.CrossEntropyLoss(weight=subweights)

print("Class Weights:")
for i, w in enumerate(subweights):
    print(f"Class {i}: weight = {w.item():.4f}")
```

```python
class_weight = [round(w.item(), 4) for w in subweights]
print(class_weight)
```

```python

import os
from PIL import Image
for split in subset_dataset:
    os.makedirs(f"yolo_subset/images/{split}", exist_ok=True)
    os.makedirs(f"yolo_subset/labels/{split}", exist_ok=True)

# Convert COCO bbox to YOLO format
def coco_to_yolo(bbox, img_w, img_h):
    x, y, w, h = bbox
    x_center = (x + w / 2) / img_w
    y_center = (y + h / 2) / img_h
    w /= img_w
    h /= img_h
    return x_center, y_center, w, h

# Process all splits
for split in subset_dataset:
    print(f"🔁 Processing split: {split}")
    for idx, sample in enumerate(subset_dataset[split]):

        img: Image.Image = sample["image"]
        img_w, img_h = img.size

        img_path = f"yolo_subset/images/{split}/{idx}.jpg"
        img.save(img_path)

        label_path = f"yolo_subset/labels/{split}/{idx}.txt"
        with open(label_path, "w") as f:
            for bbox, cls_id in zip(sample["bboxes"], sample["category_id"]):
                x, y, w, h = coco_to_yolo(bbox, img_w, img_h)
                f.write(f"{cls_id} {x:.6f} {y:.6f} {w:.6f} {h:.6f}\n")
```
